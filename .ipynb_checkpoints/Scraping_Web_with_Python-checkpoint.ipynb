## Proses Scraping

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen as uReq
```

Pada proses scraping web kali ini, saya melakukan scraping pada website Bukalapak sejumlah 7 halaman dimulai dari halaman pertama pencarian.Kata kunci yang saya gunakan adalah kata 'lipstik' dengan batasan rating dari 1 sampai dengan 5. Masing-masing halaman terdiri dari 50 data yang dapat di cek pada syntax #print(len(all_product)). Sebelum itu, cek dahulu apakah website bisa di scrape untuk di ambil datanya. Jika status koneksi = 200 maka website dapat di scrape. Data-data yang berhasil di scrape kemudian di write ke dalam format .csv agar mudah dibaca dan dianalisis kemudian. Pada proses scraping kali ini, data yang berhasil di scrap adalah sebanyak 350 data dan 6 kolom. 

```python
![ ](assets/2.png)
```

```python
csv=open("db_bukalapak.csv", 'w')

#membuat header (nama kolom) pada file db_bukalapak.csv
headers = "nama_produk,harga,penjual,kota,rating_produk,jml_review\n"
csv.write(headers)

#scraping 3 halaman yang ada di website bukalapak 
for page in range(1,8):
    #link yang akan di scrape
    base_url = 'https://www.bukalapak.com/products/s?page='+str(page)+'&search%5Bbrand_seller%5D=0&search%5Bcity%5D=&search%5Bcourier%5D=&search%5Bfree_shipping_coverage%5D=&search%5Bhashtag%5D=&search%5Binstallment%5D=0&search%5Bkeywords%5D=lipstik&search%5Bnew%5D=1&search%5Bprice_max%5D=&search%5Bprice_min%5D=&search%5Bprovince%5D=&search%5Brating_gte%5D=1&search%5Brating_lte%5D=5&search%5Btodays_deal%5D=0&search%5Btop_seller%5D=0&search%5Bused%5D=1&search%5Bwholesale%5D=0&utf8=%E2%9C%93'
    #print(base_url)

    #Request URL dan parser web 
    uClient = uReq(base_url)
    page_html = uClient.read()
    uClient.close()
    page_soup = BeautifulSoup(page_html, "html.parser")
    all_product = page_soup.find_all("div", class_ = "product-card")
    #print(len(all_product)) 
    
    #list data apa saja yang akan diambil dari produk 
    for item in all_product:
        
        # nama  dan link product
        product_name = item.find("a", {"class":"product__name line-clamp--2 js-tracker-product-link qa-list"})
        #product_link = 'https://www.bukalapak.com' + str(product_name.get('href'))
        product_name = product_name.text.replace('\n', "").strip()        

        # harga
        product_price = item.find("span", {"class":"amount positive"})
        product_price = product_price.text.replace('\n', "").strip()
        product_price = 'Rp ' + product_price        
        
        #penjual        
        seller_name = item.find("h5", {"class":"user__name"})
        seller_name = seller_name.text.replace('\n', "").strip()
        
        #kota pengiriman
        seller_city = item.find("span", {"class":"user-city__txt"})
        seller_city = seller_city.text.replace('\n', "").strip()    
        
        #rating 
        rating = item.find("div", {"class":"product__rating"})
        rating = rating.text.replace('\n', "").strip()
        rating = rating[0:3]
        
        #jumlah review
        review = item.find("div", {"class":"product__rating"})
        review = review.text.replace('\n', "").strip()
        review = review[3:]
        
        #write data yang sudah di scrape
        csv.write(product_name+','+ product_price+','+ seller_name+','+ seller_city+','+ rating+','+ review+'\n')
        #csv.write()
csv.close()
```

### Segmentasi Produk

```python
bukalapak.describe()
```

Hasil ringkasan statistika di atas akan saya gunakan untuk melakukan segmentasi produk berdasarkan harga dengan rincian sebagai berikut:
- Mahal: produk dikatakan mahal jika harganya > Q3 harga produk
- Sedang : produk dikatakan sedang jika harga >= Q2 & harga <= Q3 harga produk
- Murah : produk dikatakan murah jika harga < Q2 harga produk

```python
q_harga = (bukalapak.harga.quantile([0.25, 0.5, 0.75])).to_list()
q_harga
```

```python
kondisi = [
    (bukalapak.harga <= q_harga[1]),
    (bukalapak.harga <= q_harga[2]),
    (bukalapak.harga > q_harga[2])
]

kelas = ['Murah', 'Sedang','Mahal']
    
bukalapak['segmen_harga'] = np.select(kondisi, kelas, default='Mahal')
```

```python
bukalapak.segmen_harga.value_counts()
```

Dari hasil di atas sekarang kita dapat mengetahui bahwa setengah data penjualan lipstik yang kita miliki tergolong dalam segmen harga murah. Artinya, produk lipstik yang banyak terjual dan laku di etalase adalah produk dengan harga kurang  dari Rp 47.000,00. Untuk membuat informasi ini lebih menarik kita dapat memvisualisasikannya agar terbaca lebih mudah.

```python
bar=alt.Chart(bukalapak).mark_bar().encode(
    x = 'segmen_harga',
    y = 'count(segmen_harga)',
    color = 'segmen_harga',
    tooltip=['segmen_harga', 'count(segmen_harga)']
).interactive()
bar.properties(width=300, title='Segmentasi Harga Lipstik di Bukalapak')
```

### 10 Produk Lipstik yang Kurang Direkomendasikan

Sekarang kita akan kembali lagi ke tabel ringkasan statistika untuk merumuskan produk yang dinilai bagus dilihat dari jumlah rating dan review produknya.

```python
bukalapak.describe()
```

Berdasarkan hasil tabel diatas, saya akan membuat sebuah asumsi dengan batasan bahwa produk dikatakan kurang bagus dan kurang direkomendasikan jika rating produknya kurang dari 4.5 dan jumlah reviewnya lebih dari 3 (asumsi diambil dengan mempertimbangkan nilai quantile 1). Data yang diambil kemudian diurutkan dan diambil 10 data paling bawah.

```python
less_recomendation = bukalapak[(bukalapak.jml_review >=3)&(bukalapak.rating_produk < 4.5)].sort_values('rating_produk',ascending=False).tail(10)
```

```python
bar=alt.Chart(less_recomendation).mark_bar().encode(
    x = 'rating_produk',
    y = 'nama_produk',
    tooltip=['nama_produk','penjual','harga','jml_review','rating_produk']
)
bar.properties(width=300, title='10 Produk Lipstik yang Kurang Direkomendasikan')
```

Berdasarkan visualisasi bar plot di atas, pada bagian axis Y kita dapat melihat nama produk lipstik yang kurang direkomendasikan untuk dibeli oleh konsumen Bukalapak. Hal ini dapat menjadi pertimbangan bagi pembeli ketika membeli produk lipstik di bukalapak, atau menjadi koreksi bagi penjual untuk meningkatkan kualitas barang yang dijual di tokonya. Ada beberapa hal yang membuat sebuah produk kurang direkomendasikan yaitu kondisi barang yang tidak sama dengan yang dipajang dan dideskripsikan di etalase, barang palsu, kondisi barang yang diterima rusak, dll. Namun, ada pertimbangan yang perlu diperhatikan bahwa visualisasi ini tidak serta merta dapat menjadi acuan mengingat kualitas produk di dunia nyata sangat bergantung ke banyak hal seperti kondisi barang yang rusak akibat pengiriman dan bukan merupakan kesalahan penjual atau kurangnya wawasan pembeli sehingga menilai produk yang dibelinya adalah barang palsu, dll.

### Distribusi Data Berdasarkan Rating dan Review Produk

Seringkali kualitas sebuah produk dikaitkan dengan nilai rating dan seberapa banyak orang yang telah memberikan review terhadap produk tersebut. Hal itu kemudian menjadi pertanyaan besar apakah benar nilai rating suatu produk akan dipengaruhi oleh seberapa banyak orang yang memberikan review terhadap produk tersebut. Kali ini, saya akan melihat persebaran data penjualan lipstik sebelumnya untuk menjawab apakah pertanyaan sebelumnya terbukti atau tidak. 

```python
my_chart = alt.Chart(bukalapak).mark_point().encode(
    x='rating_produk',
    y='jml_review',
    tooltip=['nama_produk','rating_produk','jml_review']
)
my_chart.properties(width=400, title='Rating vs Jumlah Review')
```

Jika dilihat dari scatter plot di atas, data penjualan  lipstik terlihat memusat di rentang rating 4 sampai dengan 5 dengan jumlah review antara rentang 0 sampai dengan 50. Sementara itu terdapat data yang terletak di ujung kiri bawah dengan rating 1 dan jumlah review 1 yang artinya produk ini dinilai sangat jelek dan tidak direkomendasikan dan ada juga data yang terletak di ujung kanan atas dengan nilai rating 4.8 dan jumlah review sebanyak 299. Ini artinya sebenernya tidak ada kaitannya antara rating dan jumlah review pada kasus di dataset Bukalapak yang saya miliki karena dengan melihat persebaran data di atas sebenarnya walaupun jumlah review cukup banyak, rating produk tetap cukup bagus karena nilai ratingnya >= 4.5.

### Kota dengan Penjualan Terbanyak 

```python
top_10_kota = pd.crosstab(index=bukalapak.kota,
           columns='jumlah_penjualan').sort_values('jumlah_penjualan', ascending=False).head(10).reset_index()
```

```python
bars = alt.Chart(top_10_kota.reset_index()).mark_bar().encode(
    x='kota',
    y='jumlah_penjualan:Q',
    color='kota',
    tooltip=['jumlah_penjualan']
).properties(width=300)
bars
```

Jakarta Barat menjadi kota yang paling banyak melakukan pengiriman produk lipstik kepada pembeli di Bukalapak. Rata-rata penjualan produk lipstik berasal dari daerah Jabodetabek. Melalui pendekatan ini, sebuah brand lipstik dapat menjadikannya patokan untuk melakukan pendistribusian produk dengan kuantitas yang lebih besar dari sektor wilayah lainnya.
